```python
import pandas as pd

df = pd.read_csv('healthcare-dataset-stroke-data.csv')

df
```

```python
df.describe()
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
missing_percentage = (missing_values['bmi'] / len(df)) * 100
print(f"Pourcentage des valeurs manquantes dans la colonne 'bmi': {missing_percentage:.2f}%")
```

```python
# Imputation du BMI par la médiane groupée (gender + age)
df['bmi'] = df.groupby(['gender', 'age'])['bmi'].transform(
    lambda x: x.fillna(x.median())
)

# Fallback : médiane globale si des NaN restent
df['bmi'] = df['bmi'].fillna(df['bmi'].median())

# Vérification
print(f"Valeurs manquantes restantes dans bmi : {df['bmi'].isnull().sum()}")
```

```python
# Sauvegarde le nouveau CSV 
df.to_csv('healtcarecleaned.csv', index=False)
```
